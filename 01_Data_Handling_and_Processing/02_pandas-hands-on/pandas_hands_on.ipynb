## 2. Creating DataFrames

There are multiple ways to create DataFrames in pandas:
- From dictionaries
- From lists of lists
- From Series
- With custom indices

```python
import pandas as pd
import numpy as np

# Method 1: From a dictionary
print("=" * 50)
print("Creating DataFrame from Dictionary")
print("=" * 50)

data_dict = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Age': [25, 30, 35, 28],
    'City': ['New York', 'London', 'Paris', 'Tokyo'],
    'Salary': [50000, 60000, 75000, 55000]
}

df1 = pd.DataFrame(data_dict)
print(df1)
print(f"\nShape: {df1.shape}")
print(f"Data types:\n{df1.dtypes}")
```

```python
# Method 2: From a list of lists (with column names)
print("\n" + "=" * 50)
print("Creating DataFrame from List of Lists")
print("=" * 50)

data_list = [
    ['Apple', 100, 2.5],
    ['Banana', 150, 1.2],
    ['Cherry', 50, 5.0]
]

df2 = pd.DataFrame(data_list, columns=['Fruit', 'Quantity', 'Price'])
print(df2)
```

## 6. Handling Missing Data

Strategies for detecting and handling NaN values

```python
import pandas as pd
import numpy as np

# Create DataFrame with missing values
np.random.seed(42)
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5, np.nan],
    'B': [10, np.nan, np.nan, 40, 50, 60],
    'C': ['X', 'Y', 'Z', None, 'X', 'Y'],
    'D': [100, 200, 300, 400, 500, 600]
})

print("=" * 50)
print("Original DataFrame with Missing Values")
print("=" * 50)
print(df)

print("\nData types:")
print(df.dtypes)
```

```python
print("\n" + "=" * 50)
print("Detecting Missing Values")
print("=" * 50)

print("isnull():")
print(df.isnull())

print("\nSum of nulls per column:")
print(df.isnull().sum())
```

```python
print("\n" + "=" * 50)
print("Dropping Missing Values")
print("=" * 50)

# Drop rows with any NaN
print("After dropna():")
print(df.dropna())

print("\nAfter dropna(subset=['A']):")
print(df.dropna(subset=['A']))
```

```python
print("\n" + "=" * 50)
print("Filling Missing Values")
print("=" * 50)

# Fill with constant value
print("fillna with constant:")
print(df.fillna(0))

print("\nfillna with forward fill:")
print(df.ffill())
```

```python
print("\nfillna with backward fill:")
print(df.bfill())

print("\nfillna with interpolation:")
df_interp = df.copy()
df_interp['A'] = df_interp['A'].interpolate()
print(df_interp)
```

```python
print("\n" + "=" * 50)
print("Advanced: Missing Data Indicators")
print("=" * 50)

# Add indicator for missing values
result = df.copy()
result['has_missing_A'] = result['A'].isnull()
print(result)
```

## 7. Data Transformation

Creating new columns, conditional logic, and data type conversions

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [25, 30, 35, 28, 32],
    'Salary': [50000, 60000, 75000, 55000, 65000],
    'Department': ['Sales', 'Engineering', 'Sales', 'HR', 'Engineering']
})

print("=" * 50)
print("Creating New Columns")
print("=" * 50)

df['Annual_Bonus'] = df['Salary'] * 0.1
print(df)
```

```python
print("\n" + "=" * 50)
print("Conditional Columns with np.where()")
print("=" * 50)

df['Bonus_Level'] = np.where(df['Salary'] > 60000, 'High', 'Standard')
print(df[['Name', 'Salary', 'Bonus_Level']])
```

```python
print("\n" + "=" * 50)
print("Complex Conditions with np.select()")
print("=" * 50)

conditions = [
    df['Salary'] > 70000,
    (df['Salary'] > 55000) & (df['Salary'] <= 70000),
    df['Salary'] <= 55000
]
choices = ['Premium', 'Standard', 'Basic']

df['Tier'] = np.select(conditions, choices, default='Unknown')
print(df[['Name', 'Salary', 'Tier']])
```

```python
print("\n" + "=" * 50)
print("Using apply() for Custom Functions")
print("=" * 50)

# Apply to entire column
df['Age_Group'] = df['Age'].apply(lambda x: 'Young' if x < 30 else 'Senior')
print(df[['Name', 'Age', 'Age_Group']])

# Apply with custom function
def salary_category(salary):
    if salary < 55000:
        return 'Entry-Level'
    elif salary < 65000:
        return 'Mid-Level'
    else:
        return 'Senior'

df['Salary_Category'] = df['Salary'].apply(salary_category)
print(df[['Name', 'Salary', 'Salary_Category']])
```

```python
print("\n" + "=" * 50)
print("Renaming Columns")
print("=" * 50)

df_renamed = df.rename(columns={
    'ID': 'Employee_ID',
    'Name': 'Employee_Name',
    'Annual_Bonus': 'Bonus'
})
print(df_renamed.columns.tolist())
```

```python
print("\n" + "=" * 50)
print("Sorting Data")
print("=" * 50)

# Sort by single column
print("Sorted by Salary (ascending):")
print(df.sort_values('Salary'))

print("\nSorted by Age (descending):")
print(df.sort_values('Age', ascending=False))

print("\nSorted by multiple columns:")
print(df.sort_values(['Department', 'Salary'], ascending=[True, False]))
```

```python
print("\n" + "=" * 50)
print("Type Conversion")
print("=" * 50)

df['Age_str'] = df['Age'].astype(str)
print(f"Age converted to string: {df['Age_str'].dtype}")

df['Salary_float'] = df['Salary'].astype(float)
print(f"Salary as float: {df['Salary_float'].dtype}")
```

```python
print("\n" + "=" * 50)
print("Binning Data with pd.cut() and pd.qcut()")
print("=" * 50)

# Equal-width bins
df['Age_Bin_cut'] = pd.cut(df['Age'], bins=3, labels=['Young', 'Middle', 'Old'])
print("pd.cut (equal-width bins):")
print(df[['Name', 'Age', 'Age_Bin_cut']])

# Equal-frequency bins
df['Age_Bin_qcut'] = pd.qcut(df['Age'], q=2, labels=['Lower', 'Upper'])
print("\npd.qcut (equal-frequency bins):")
print(df[['Name', 'Age', 'Age_Bin_qcut']])
```

```python
print("\n" + "=" * 50)
print("Drop Columns")
print("=" * 50)

cols_to_drop = ['Age_str', 'Salary_float', 'Age_Bin_cut', 'Age_Bin_qcut', 'Bonus_Level', 'Tier', 'Age_Group', 'Salary_Category', 'Annual_Bonus']
df_dropped = df.drop(cols_to_drop, axis=1)
print(df_dropped.columns.tolist())
```

## 8. Grouping & Aggregation

Using groupby for data summarization

```python
import pandas as pd
import numpy as np

np.random.seed(42)
df = pd.DataFrame({
    'Department': ['Sales', 'Engineering', 'Sales', 'Engineering', 'HR', 'Sales', 'Engineering', 'HR'],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank', 'Grace', 'Henry'],
    'Salary': [50000, 75000, 55000, 80000, 45000, 52000, 78000, 48000],
    'Years': [2, 5, 3, 6, 1, 2, 7, 1]
})

print("=" * 50)
print("Basic GroupBy")
print("=" * 50)

print("Group by Department:")
grouped = df.groupby('Department')
print(grouped['Salary'].sum())
```

```python
print("\n" + "=" * 50)
print("Multiple Aggregations")
print("=" * 50)

result = df.groupby('Department')['Salary'].agg(['sum', 'mean', 'min', 'max', 'count'])
print(result)
```

```python
print("\n" + "=" * 50)
print("Named Aggregations")
print("=" * 50)

result_named = df.groupby('Department').agg(
    Total_Salary=('Salary', 'sum'),
    Avg_Salary=('Salary', 'mean'),
    Employee_Count=('Name', 'count'),
    Max_Salary=('Salary', 'max')
)
print(result_named)
```

```python
print("\n" + "=" * 50)
print("GroupBy with Transform")
print("=" * 50)

# Add department average salary as new column
df['Dept_Avg_Salary'] = df.groupby('Department')['Salary'].transform('mean')
print(df[['Name', 'Department', 'Salary', 'Dept_Avg_Salary']])
```

```python
print("\n" + "=" * 50)
print("Filtering Groups")
print("=" * 50)

# Filter groups with more than 2 employees
large_groups = df.groupby('Department').filter(lambda x: len(x) > 2)
print("Departments with more than 2 employees:")
print(large_groups)
```

## 9. Merging & Joining DataFrames

Combining DataFrames with various join strategies

```python
import pandas as pd

# Create sample datasets
employees = pd.DataFrame({
    'Emp_ID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Dept_ID': [10, 20, 10, 30, 20]
})

departments = pd.DataFrame({
    'Dept_ID': [10, 20, 30, 40],
    'Dept_Name': ['Sales', 'Engineering', 'HR', 'Marketing']
})

print("=" * 50)
print("Employees DataFrame:")
print(employees)
print("\nDepartments DataFrame:")
print(departments)
```

```python
print("\n" + "=" * 50)
print("Inner Merge")
print("=" * 50)

inner_merge = pd.merge(employees, departments, on='Dept_ID', how='inner')
print(inner_merge)
```

```python
print("\n" + "=" * 50)
print("Left Merge")
print("=" * 50)

left_merge = pd.merge(employees, departments, on='Dept_ID', how='left')
print(left_merge)
```

```python
print("\n" + "=" * 50)
print("Right Merge")
print("=" * 50)

right_merge = pd.merge(employees, departments, on='Dept_ID', how='right')
print(right_merge)
```

```python
print("\n" + "=" * 50)
print("Outer Merge")
print("=" * 50)

outer_merge = pd.merge(employees, departments, on='Dept_ID', how='outer')
print(outer_merge)
```

```python
print("\n" + "=" * 50)
print("Merge with Indicator")
print("=" * 50)

merge_indicator = pd.merge(employees, departments, on='Dept_ID',
                           how='outer', indicator=True)
print(merge_indicator)
```

```python
print("\n" + "=" * 50)
print("Concatenation")
print("=" * 50)

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

print("Concatenate vertically (rows):")
concat_result = pd.concat([df1, df2], ignore_index=True)
print(concat_result)

print("\nConcatenate horizontally (columns):")
concat_h = pd.concat([df1, df2], axis=1)
print(concat_h)
```

## 10. String Operations

Text manipulation with pandas .str accessor

```python
import pandas as pd

df = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Email': ['  alice@example.com  ', 'bob@example.com', 'CHARLIE@EXAMPLE.COM', 'diana@example.com'],
    'Address': ['123 Main St, NYC', '456 Oak Ave, LA', '789 Pine Rd, CHI', '321 Elm Blvd, BOS']
})

print("=" * 50)
print("String Methods")
print("=" * 50)

print("Original:")
print(df['Email'])

print("\nAfter strip():")
print(df['Email'].str.strip())

print("\nAfter lower():")
print(df['Email'].str.lower())

print("\nAfter upper():")
print(df['Email'].str.upper())
```

## 11. Date & Time Operations

Working with dates and time series data

```python
import pandas as pd
import numpy as np

# Create date range
dates = pd.date_range('2024-01-01', periods=10, freq='D')
df = pd.DataFrame({
    'Date': dates,
    'Value': np.random.randint(50, 200, 10)
})

print("=" * 50)
print("Date Operations")
print("=" * 50)

print("Original DataFrame:")
print(df)

print("\nData type of Date column:")
print(df['Date'].dtype)
```

```python
print("\n" + "=" * 50)
print("Date Accessors")
print("=" * 50)

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['DayName'] = df['Date'].dt.day_name()
df['Quarter'] = df['Date'].dt.quarter
df['WeekOfYear'] = df['Date'].dt.isocalendar().week

print(df[['Date', 'Year', 'Month', 'Day', 'DayName', 'Quarter', 'WeekOfYear']])
```

```python
print("\n" + "=" * 50)
print("Date Arithmetic")
print("=" * 50)

df2 = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=5)
})

df2['Next_Week'] = df2['Date'] + pd.Timedelta(weeks=1)
df2['Last_Month'] = df2['Date'] - pd.DateOffset(months=1)

print(df2)
```

```python
print("\n" + "=" * 50)
print("Rolling Windows")
print("=" * 50)

df3 = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=10),
    'Sales': [100, 150, 120, 180, 160, 200, 190, 210, 220, 240]
})

df3['Rolling_Mean_3'] = df3['Sales'].rolling(window=3).mean()
df3['Rolling_Sum_2'] = df3['Sales'].rolling(window=2).sum()

print(df3)
```

```python
print("\n" + "=" * 50)
print("Resample Time Series")
print("=" * 50)

df4 = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=30, freq='D'),
    'Sales': np.random.randint(100, 500, 30)
})

df4.set_index('Date', inplace=True)

print("Weekly aggregation (sum):")
weekly_sum = df4.resample('W').sum()
print(weekly_sum.head())

print("\nWeekly aggregation (mean):")
weekly_mean = df4.resample('W').mean()
print(weekly_mean.head())
```

## 12. Pivot Tables & Reshaping

Reshaping and summarizing data with pivot operations

```python
import pandas as pd
import numpy as np

# Create sample data
df = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=12),
    'Product': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Region': ['North', 'South', 'North', 'South', 'East', 'West', 'East', 'West', 'North', 'South', 'East', 'West'],
    'Sales': [100, 150, 120, 180, 90, 110, 140, 160, 130, 200, 95, 125]
})

print("=" * 50)
print("Sample Data:")
print(df)
```

```python
print("\n" + "=" * 50)
print("Pivot Table")
print("=" * 50)

pivot_result = df.pivot_table(
    values='Sales',
    index='Product',
    columns='Region',
    aggfunc='sum',
    fill_value=0
)
print(pivot_result)
```

```python
print("\n" + "=" * 50)
print("Melt (Unpivot)")
print("=" * 50)

# Create a wide DataFrame
wide_df = pd.DataFrame({
    'Product': ['A', 'B', 'C'],
    'Q1': [100, 120, 90],
    'Q2': [150, 140, 110],
    'Q3': [120, 160, 130]
})

print("Wide format:")
print(wide_df)

print("\nMelted (long format):")
melted = wide_df.melt(id_vars=['Product'], var_name='Quarter', value_name='Sales')
print(melted)
```

```python
print("\n" + "=" * 50)
print("Crosstab")
print("=" * 50)

crosstab_result = pd.crosstab(
    index=df['Product'],
    columns=df['Region'],
    values=df['Sales'],
    aggfunc='sum'
)
print(crosstab_result)
```

## 13. Performance Tips & Optimization

Strategies for efficient pandas operations

```python
import pandas as pd
import numpy as np
import time

# Create large dataset
np.random.seed(42)
df = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C', 'D'], 10000),
    'Value': np.random.randint(0, 100, 10000),
    'Amount': np.random.uniform(0, 1000, 10000)
})

print("=" * 50)
print("Performance: Category Dtype")
print("=" * 50)

# Before optimization
print(f"Memory before: {df['Category'].memory_usage(deep=True) / 1024:.2f} KB")

# After optimization
df['Category_cat'] = df['Category'].astype('category')
print(f"Memory after: {df['Category_cat'].memory_usage(deep=True) / 1024:.2f} KB")
```

```python
print("\n" + "=" * 50)
print("Performance: Vectorization vs Loop")
print("=" * 50)

test_df = pd.DataFrame({
    'A': np.random.randint(0, 100, 100),
    'B': np.random.randint(0, 100, 100)
})

# Using loop (slow)
start = time.time()
result_loop = []
for i in range(len(test_df)):
    result_loop.append(test_df.loc[i, 'A'] + test_df.loc[i, 'B'])
loop_time = time.time() - start

# Using vectorization (fast)
start = time.time()
result_vec = test_df['A'] + test_df['B']
vec_time = time.time() - start

print(f"Loop time: {loop_time:.6f} seconds")
print(f"Vectorized time: {vec_time:.6f} seconds")
print(f"Speedup: {loop_time/vec_time:.1f}x faster")
```

```python
print("\n" + "=" * 50)
print("Method Chaining")
print("=" * 50)

# Create sample data
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Age': [25, 30, 35, 28],
    'Salary': [50000, 60000, 75000, 55000]
})

# Method chaining - clean and efficient
result = (df
    .query('Age > 26')
    .assign(Bonus=lambda x: x['Salary'] * 0.1)
    .sort_values('Salary', ascending=False)
    [['Name', 'Salary', 'Bonus']]
)

print(result)
```
